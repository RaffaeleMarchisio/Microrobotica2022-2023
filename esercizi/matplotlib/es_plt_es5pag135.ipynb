```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_star=100
x=np.random.uniform(0.,1.,n_star)
y=np.random.uniform(0.,1.,n_star)
```

```python
fig,ax=plt.subplots(figsize=(10,10))
ax.set_facecolor("black")
ax.plot(x,y,marker="*",markersize=20, linestyle="-",color="white")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
x=np.linspace(0,4*np.pi,100)
s1=np.sin(x)
s2=np.cos(x)
```

```python
fig,ax=plt.subplots(figsize=(14,7))
ax.plot(x,s1,linestyle=":",linewidth=3,color="red", label="sin")
ax.plot(x,s2,linestyle="-.",linewidth=3,color="green", label="cos")
ax.set_xlabel("x",fontsize=30)
ax.set_ylabel("y",fontsize=30)
ax.tick_params(axis="both",labelsize=20)
ax.set_title("funzioni oscillanti",fontsize=30)
ax.legend()
ax.grid()
plt.savefig("grafico.png",bbox_inches="tight")
plt.show()
```

# grafici a dispersione

```python
n_star=100
x=np.random.uniform(0.,1.,n_star)
y=np.random.uniform(0.,1.,n_star)
colori=np.random.uniform(0.,1.,(n_star,3))
dim=np.random.uniform(10.,500.,n_star)
trasparenza=np.random.uniform(0.,1.,n_star)
```

```python
fig,ax=plt.subplots(figsize=(4,4))
ax.scatter(x,y,c=colori, marker="*", s=dim, alpha = trasparenza)
plt.show()
```

```python
scelte=np.random.choice(("est","ovest","sud","nord"),7200)
scelte
x=0
y=0

posx=[]
posy=[]
```

```python
i=0
for valori in scelte:
    if valori=="nord":
        y=y+1
        posy.append(y);
    elif valori=="sud":
        y=y-1
        posy.append(y);
    elif valori=="est":
        x=x+1
        posx.append(x);
    elif valori=="ovest":
        x=x-1
        posx.append(x);
```

```python
fig,ax=plt.subplots(figsize=(10,10))
ax.plot(0,0, marker="*",label="START",markersize=15,color="lightgreen")
while len(posx)<7200:
    posx.append(posx[len(posx)-1])
while len(posy)<7200:
    posy.append(posy[len(posy)-1])
ax.plot(posx,posy,linestyle="--",color="white")
ax.plot(posx[len(posx)-1],posy[len(posy)-1],marker="*", label="END",markersize=15,color="yellow")
ax.legend()
ax.set_facecolor("black")
ax.set_title("Robot Pazzo")
ax.set_xlabel("X",fontsize=20)
ax.set_ylabel("Y",fontsize=20)
plt.show()
posx[len(posx)-2]
```
