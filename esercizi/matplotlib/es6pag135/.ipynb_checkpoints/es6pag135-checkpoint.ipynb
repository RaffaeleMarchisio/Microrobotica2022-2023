```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
righe=[]
```

```python

anni = []
valori = []
with open("popolazioneMondiale.csv", "r") as file:
    righe = csv.reader(file)
    next(righe)
    for row in righe:
        if row[0]!="Anno ":
            anni.append(int(row[0]))
            valori.append(int(row[1]))





```

```python

x = anni
y = valori

fig, ax = plt.subplots(figsize = (20,10))
ax.plot(x,y)
ax.set_xlabel("ANNI", fontsize=30)
ax.set_ylabel("POPOLAZIONE", fontsize=30)

plt.show()

anni
```
