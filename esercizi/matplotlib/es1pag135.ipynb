```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,5)
F1=x
F2=x*x
F3=x*x*x
F3
```

```python
plt.plot(x,F1,label="funzione x")
plt.plot(x,F2,label="funzione x^2")
plt.plot(x,F3,label="Funzione x^3")
plt.xlabel("ascissa")
plt.ylabel("ordinata")
plt.legend()
plt.show
```
