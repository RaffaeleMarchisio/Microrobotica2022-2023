```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-1,1,200)
F1=x**2
F2=x**3
F3=x**4
F3
```

```python
fig,ax=plt.subplots(figsize=(10,5))
ax.plot(x,F1,label="funzione x")
ax.plot(x,F2,label="funzione x^2")
ax.plot(x,F3,label="Funzione x^3")
ax.set_xlabel("ascissa")
ax.set_ylabel("ordinata")
plt.title("Grafico delle funzioni")
ax.legend()
plt.show()
```
