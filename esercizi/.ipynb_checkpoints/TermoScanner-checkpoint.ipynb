```python
import numpy as np
import uuid

n_passegeri=200000
temperature=np.random.normal(36.5,0.25,n_passegeri)
identificativi=np.array([str(uuid.uuid4()) for _ in range(n_passegeri)])
print(temperature)
print(identificativi)
```

# SOLUZIONE CON PYTHON

```python
%%timeit
#1) il numero totale di passeggeri con temperatura maggiore di 37.5 °C.
contatore=0

for t in temperature:
    if t>37.5:
        contatore+=1
print(f"il numero di passegeri con temperatura magggiore di 37.5 °C è {contatore}\n")

#2)tutti gli identificativi dei passeggeri con temperatura maggiore di 37.5°C.

print("ATTENZIONE:Passeggeri con temperatura maggiore di 37.5 °C:")

for i in range(n_passegeri):
    if temperature[i]>37.5:
        print(identificativi[i])
```

# SOLUZIONE CON NUMPY

```python
%%timeit
#1) il numero totale di passeggeri con temperatura maggiore di 37.5 °C.ù
cont=np.sum(temperature > 37.5)
print(f"il numero di passegeri con temperatura magggiore di 37.5 °C è {cont}")

#2)tutti gli identificativi dei passeggeri con temperatura maggiore di 37.5°C.
print("ATTENZIONE:Passeggeri con temperatura maggiore di 37.5 °C:")
print(identificativi[temperature>37.5])
```
