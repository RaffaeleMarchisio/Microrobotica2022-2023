```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
voti10=np.floor(np.random.normal(6,3,100))
voti10[voti10<2]=2
voti10[voti10>10]=10
voti10
```

```python
voti30=np.floor(np.random.normal(18,9,100))
voti30[voti30<2]=2
voti30[voti30>30]=30
voti30
```

```python
fig,axs=plt.subplots(nrows=2,figsize=(5,5))
axs[0].hist(voti10,bins=10)
axs[1].hist(voti30,bins=30)
plt.show()
```

```python
voti10normalizzati=(voti10-np.mean(voti10))/np.std(voti10)
voti30normalizzati=(voti30-np.mean(voti30))/np.std(voti30)
print(voti10normalizzati)
voti30normalizzati
```

```python
fig,axs=plt.subplots(nrows=2,figsize=(5,5))
axs[0].hist(voti10normalizzati,bins=8)
axs[1].hist(voti30normalizzati,bins=8)
plt.show()
```
