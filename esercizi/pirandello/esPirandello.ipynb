```python
from collections import Counter
```

```python
f=open("sei_personaggi_in_cerca_dautore_Luigi_Pirandello.txt","r")
righe = f.readlines()
stringa = "".join(righe)
strminuscolo = stringa.lower()
```

```python
listapulita = [",", ";", ".", ":", "!", "?", "(", ")", "'","-","»","«","..."]
listapulitatot  = [lettera for lettera in strminuscolo if lettera not in listapulita]
stringapulita = "".join(listapulitatot)
stringa_split2 = stringapulita.split()
dic = Counter(stringa_split2) 
print(dic)
f.close()
```

```python
d={}
for parola in stringa_split2:
    if parola not in d:
        d[parola]=[stringa_split2[i+1] for i,par in enumerate(stringa_split2[:-1]) if parola==par]
d
```
