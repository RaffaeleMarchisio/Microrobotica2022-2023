```python
df=pd.read_csv("penguins_size.csv")
df
```

```python
colonne=["culmen_length_mm","culmen_depth_mm","flipper_length_mm","body_mass_g"]
colonne
```

```python
fig,axs=plt.subplots(nrows=4,ncols=4,figsize=(10,10))
for i, colonna1 in enumerate(colonne):
    for j,colonna2 in enumerate(colonne):
        if i==j:
            pass
        else:
            var_x=df[colonna1].values
            var_y=df[colonna2].values
            axs[i,j].scatter(var_x,var_y,s=10)
plt.show()
```
