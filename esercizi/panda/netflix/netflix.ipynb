```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
netflix=pd.read_csv("netflix.csv")
netflix
```

```python
fig,ax=plt.subplots(figsize=(10,10))
ax.scatter(netflix["Awards Received"], netflix["IMDb Score"], s=netflix["IMDb Votes"]/10000)
ax.set_xlabel("Premi ricevuti")
ax.set_ylabel("Voto")
plt.show()
```
