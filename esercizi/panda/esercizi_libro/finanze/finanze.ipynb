```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("data_sp500.csv")
df
```

```python
df.info()
```

```python
fig,ax=plt.subplots(figsize=(10,10))
df["SP500"].plot(ax=ax)
plt.show()
```

```python
df["SP500_return"]=(df['SP500']-df['SP500'].shift(1))/df["SP500"].shift(1)
```

```python
fig,ax=plt.subplots(figsize=(10,6))
df["SP500_return"].plot(ax=ax)
plt.show()
```

# BITCOIN

```python
bitdf=pd.read_csv("bitcoin_csv.csv",index_col="date")
bitdf
```

```python
bitdf.dropna(inplace=True)
bitdf
```

```python
fig,ax=plt.subplots(figsize=(13,5))
bitdf["price(USD)"].plot(ax=ax)
plt.show()
```

```python
bitdf["prezzo_return"]=(bitdf['price(USD)']-bitdf['price(USD)'].shift(1))/bitdf['price(USD)'].shift(1)
bitdf["prezzo_return"]
```

```python
fig,ax=plt.subplots(figsize=(10,6))
bitdf["prezzo_return"].plot(ax=ax)
plt.show()
```

# FELICITA

```python
happy2015=pd.read_csv("2015.csv")
happy2016=pd.read_csv("2016.csv")
happy2017=pd.read_csv("2017.csv")
happy2018=pd.read_csv("2018.csv")
happy2019=pd.read_csv("2019.csv")
happy=pd.concat([happy2015,happy2016,happy2017,happy2018,happy2019])
happy
```

```python
happy_perrichhezza=happy[["Happiness Score","Economy (GDP per Capita)","Country"]].dropna()
happy_perrichhezza
```

```python
countries=happy_perrichhezza["Country"].values
x=happy_perrichhezza["Happiness Score"].values
y=happy_perrichhezza["Economy (GDP per Capita)"].values
```

```python
fig,ax=plt.subplots(figsize=(18,18))
ax.scatter(happy_perrichhezza["Happiness Score"],happy_perrichhezza["Economy (GDP per Capita)"])
for posx,posy,C in zip(x,y,countries):
    ax.annotate(C,(posx,posy), fontsize=8)
ax.set_ylim([0.75,1.75])
ax.set_xlim([5,8])
plt.show()
```
