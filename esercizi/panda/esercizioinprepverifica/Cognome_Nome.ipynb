2) Quali sono i 5 paesi che hanno vinto più medaglie?

```python
top5=dfmed[["country","medal_type"]].groupby(["country"]).count().sort_values(by=["medal_code"],ascending=False)[0:5]
top5
```

```python
top5medtype=dfmed[["country","medal_type","medal_code"]].groupby(["country","medal_type"]).count()
top5medtype=top5medtype.reset_index()
top5medtype
```

```python
top5medtype=top5medtype[top5medtype['country'].isin(top5.index)]
top5medtype
```

```python
top5medtype=top5medtype.set_index(["country","medal_type"])
top5medtype.plot.bar()
```

```python
atl_ita_disc=dfatl[dfatl["country_code"]=="ITA"].groupby("discipline").count().drop(["name","gender","birth_date","birth_place","birth_country","country","country_code","discipline_code","residence_place","residence_country","height_m/ft","url"],axis=1)
atl_ita_disc
```

```python
atl_ita_disc.plot.bar().set_ylabel("N_atleti_italiani")
```

```python
grafico=dfNetflix[["IMDb Score","Awards Received","IMDb Votes"]].dropna()
grafico
```

```python
fig,ax=plt.subplots(figsize=(20,20))
ax.scatter(grafico["IMDb Score"].values,grafico['Awards Received'].values,c=grafico['IMDb Votes'].values,cmap="Reds",s=grafico["IMDb Votes"]/5000,alpha=1)
ax.grid()
plt.show()
```
