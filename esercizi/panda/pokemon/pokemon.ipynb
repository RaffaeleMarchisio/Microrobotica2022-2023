```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
data= pd.read_csv("pokemon.csv")
data
```

```python
data.describe()
data = data.replace(np.nan,"NaN")
```

```python
data["Name"].values
```

```python
np.random.choice(data["Name"].values,5)
```

```python

pokemon_im=[]
for elemento in range(len(data["Name"])):
    try:
        pokemon_im.append(plt.imread(f"images/{data['Name'][elemento]}.png"))

    except:
        pokemon_im.append(plt.imread(f"images/{data['Name'][elemento]}.jpg"))
pokemon_im
pokemon_im
```

```python
pokemon_im_noalpha=[]
for elemento in pokemon_im:
    pokemon_im_noalpha.append(elemento[:,:,0:3])
pokemon_im_noalpha=np.array(pokemon_im_noalpha)
```

```python

ired=[]
igreen=[]
iblue=[]
for img in pokemon_im_noalpha:
    if np.max(img)>1:
        img=img/255
    ired.append(np.mean(img[:,:,0]))
    igreen.append(np.mean(img[:,:,1]))
    iblue.append(np.mean(img[:,:,2]))
ired
```

```python
data["RedIntensity"]=ired
data["BlueIntensity"]=iblue
data["GrennIntensity"]=igreen
data
```

```python
data_agg = data[["Type1","RedIntensity","GrennIntensity","BlueIntensity"]].groupby("Type1").mean()
```

```python
array_red=data_agg["RedIntensity"].values
array_green=data_agg["GrennIntensity"].values
array_blue=data_agg["BlueIntensity"].values
array_index=data_agg.index
```

```python
fig,ax=plt.subplots(figsize=(10,3))
ax.plot(array_index,array_red,label="red")
ax.plot(array_index,array_green,label="green")
ax.plot(array_index,array_blue,label="blue")
plt.xticks(rotation=45, ha='right')
ax.legend()
plt.show()
```

```python
type1="Normal"
type2="Flying"
pokemon=data[(data["Type1"]==type1) & (data["Type2"]==type2)]
indexes=pokemon.index
indexes
```

```python
fig,axs=plt.subplots(nrows=1+len(pokemon)//4, ncols=4, figsize=(10,10))
cont=0
for i in range(len(axs)):
    for j in range(len(axs)):
        axs[i,j].imshow(pokemon_im[indexes[cont]])
        axs[i,j].set_axis_off()
        cont=cont+1
        if cont==len(pokemon):
            break
```
